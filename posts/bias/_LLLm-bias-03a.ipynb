```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial import Voronoi, voronoi_plot_2d, cKDTree

```

```python
# Define your points, for example, 5 random points
points = np.array([[2, 3], [3, 8], [4, 5], [6, 7], [7, 2], [3,1], [9,4]])

# Create the Voronoi diagram
vor = Voronoi(points)

# Plot the Voronoi diagram
fig, ax = plt.subplots()
voronoi_plot_2d(vor, ax=ax, show_vertices=False, line_colors='orange', line_width=2, line_alpha=0.6, point_size=10)
ax.plot(points[:,0], points[:,1], 'b.', markersize=10)

# New point
new_point = np.array([5, 5])

# Create a KDTree for efficient nearest neighbor search
tree = cKDTree(points)

# Find the nearest original point (Voronoi site) to the new point
_, index = tree.query(new_point)

# Plot the new point
ax.plot(new_point[0], new_point[1], 'r*', markersize=12)  # Plot new point in red

# Highlight the Voronoi cell the new point falls into by plotting a circle around the nearest Voronoi site
nearest_voronoi_site = points[index]
#circle = plt.Circle(nearest_voronoi_site, 0.5, color='green', fill=False, linewidth=2)
#ax.add_artist(circle)

plt.show()
```
